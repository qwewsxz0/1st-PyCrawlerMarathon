# 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
# API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 
# {"error":1106,"message":"\"popular\" must be one of [true, on, false, normalized, trending]","field":"popular"}
# API: https://www.dcard.tw/_api/forums/pet/posts?popular=false 

```python
import json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text
data = json.loads(response)
likeCount_sum_pop = 0
commentCount_sum_pop = 0
for d in data:
    print('「標題」: ',d['title'])
    print('「貼文時間」: ',d['createdAt'])
    print('「留言人數」: ',d['commentCount'])
    print('「按讚人數」: ',d['likeCount'])
    likeCount_sum_pop = likeCount_sum_pop + d['likeCount']
    commentCount_sum_pop = commentCount_sum_pop + d['commentCount']
```

```python
likeCount_sum_pop/len(data)
```

```python
commentCount_sum_pop/len(data)
```

```python
import json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response = r.text
data = json.loads(response)
likeCount_sum_nonpop = 0
commentCount_sum_nonpop = 0
for d in data:
    print('「標題」: ',d['title'])
    print('「貼文時間」: ',d['createdAt'])
    print('「留言人數」: ',d['commentCount'])
    print('「按讚人數」: ',d['likeCount'])
    likeCount_sum_nonpop = likeCount_sum_nonpop + d['likeCount']
    commentCount_sum_nonpop = commentCount_sum_nonpop + d['commentCount']
```

```python
likeCount_sum_nonpop/len(data)
```

```python
commentCount_sum_nonpop/len(data)
```
